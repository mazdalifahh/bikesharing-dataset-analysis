### Assessing Data

```python
# Cek Informasi Dataset
print("\nInformasi Dataset Harian (day.csv):")
print(df_day.info())

print("\nInformasi Dataset Per Jam (hour.csv):")
print(df_hour.info())

# Cek Jumlah Duplikasi
dup_day = df_day.duplicated().sum()
dup_hour = df_hour.duplicated().sum()

print("\nCek Data Duplikat:")
print(f"Duplikasi di day.csv: {dup_day} {'(Tidak ada duplikasi)' if dup_day == 0 else '(Ada duplikasi!)'}")
print(f"Duplikasi di hour.csv: {dup_hour} {'(Tidak ada duplikasi)' if dup_hour == 0 else '(Ada duplikasi!)'}")

# Cek Missing Values
missing_day = df_day.isnull().sum()
missing_hour = df_hour.isnull().sum()

print("\nCek Missing Values:")
if missing_day.any():
    print(f"Missing values di day.csv:\n{missing_day[missing_day > 0]}")
else:
    print("Tidak ada missing values di day.csv.")

if missing_hour.any():
    print(f"Missing values di hour.csv:\n{missing_hour[missing_hour > 0]}")
else:
    print("Tidak ada missing values di hour.csv.")

# Fungsi deteksi outlier dengan IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return len(outliers), (len(outliers) / len(data)) * 100

# Deteksi Outlier
columns_to_check = ["cnt", "temp", "hum", "windspeed"]

print("\nDeteksi Outlier (Metode IQR):")
for col in columns_to_check:
    num_outliers, perc_outliers = detect_outliers_iqr(df_day, col)
    print(f"Outlier di '{col}': {num_outliers} data ({perc_outliers:.2f}%)")
```

**Insight:**
- Data Lengkap: Tidak ada nilai yang hilang (missing values) pada kedua dataset, day.csv dan hour.csv, sehingga siap untuk analisis lebih lanjut.
- Tidak Ada Duplikasi: Tidak ada duplikasi data pada kedua dataset, yang memastikan data sudah bersih dari duplikasi yang bisa mempengaruhi hasil analisis.
- Outlier:
Kolom cnt (jumlah sepeda) dan temp (suhu) tidak memiliki outlier, yang menunjukkan data dalam rentang yang normal.
Kolom hum (kelembapan) memiliki 2 outlier, dan windspeed memiliki 13 outlier, yang mungkin perlu diperiksa lebih lanjut apakah itu data yang valid atau perlu dibersihkan.
- Memori dan Ukuran Data: Dataset hour.csv lebih besar dengan 17.379 baris, sementara day.csv hanya 731 baris, yang memberikan informasi per jam dan per hari.


### Cleaning Data

```python
#Tidak ada duplikat

#Tidak ada missing values

#Konversi tipe data
# Konversi dteday ke datetime
df_day['dteday'] = pd.to_datetime(df_day['dteday'])
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])

# Ubah kolom kategorikal
categorical_cols_day = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
categorical_cols_hour = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']

for col in categorical_cols_day:
    df_day[col] = df_day[col].astype('category')

for col in categorical_cols_hour:
    df_hour[col] = df_hour[col].astype('category')
```

```python
# Menampilkan Baris yang Mengandung Outlier

# Menentukan batas outlier menggunakan IQR
Q1_hum = df_day['hum'].quantile(0.25)
Q3_hum = df_day['hum'].quantile(0.75)
IQR_hum = Q3_hum - Q1_hum
lower_bound_hum = Q1_hum - 1.5 * IQR_hum
upper_bound_hum = Q3_hum + 1.5 * IQR_hum

Q1_windspeed = df_day['windspeed'].quantile(0.25)
Q3_windspeed = df_day['windspeed'].quantile(0.75)
IQR_windspeed = Q3_windspeed - Q1_windspeed
lower_bound_windspeed = Q1_windspeed - 1.5 * IQR_windspeed
upper_bound_windspeed = Q3_windspeed + 1.5 * IQR_windspeed

# Filter outlier
outlier_hum = df_day[(df_day['hum'] < lower_bound_hum) | (df_day['hum'] > upper_bound_hum)]
outlier_windspeed = df_day[(df_day['windspeed'] < lower_bound_windspeed) | (df_day['windspeed'] > upper_bound_windspeed)]

# Menampilkan hasil
print("Outlier pada Humidity (hum):")
display(outlier_hum)

print("\nOutlier pada Windspeed:")
display(outlier_windspeed)

# (hum dan windspeed masih dalam rentang wajar, jadi tidak perlu dihapus atau dirubah)
```

**Insight:**
- Outlier pada Humidity (hum): Dua nilai yang rendah (0.000 dan 0.1879) tetap dibiarkan, karena meskipun tidak biasa, nilai ini masih dalam rentang yang mungkin terjadi dalam situasi tertentu, seperti saat cuaca sangat kering atau pengukuran yang sangat spesifik.
- Outlier pada Windspeed (windspeed): Beberapa nilai tinggi pada windspeed tetap dibiarkan, karena angin kencang bisa terjadi dalam kondisi cuaca tertentu, dan nilai-nilai ini tetap dalam rentang yang dapat diterima jika melihat variasi cuaca.
- Langkah Selanjutnya: Dengan membiarkan outlier tersebut, kita bisa melanjutkan analisis dengan fokus pada hubungan antara variabel lain tanpa khawatir tentang pengaruh yang signifikan dari outlier, karena dalam konteks data ini, nilai-nilai ekstrem masih masuk akal dalam variasi cuaca dan kondisi nyata.

## Exploratory Data Analysis (EDA) 

```python
import pandas as pd

# Format angka agar lebih mudah dibaca
pd.options.display.float_format = '{:.2f}'.format

# Statistik deskriptif untuk df_main (dataset harian)
print("\nStatistik Deskriptif Dataset Harian (df_main):")
display(df_main.describe())

# Statistik deskriptif untuk df_hour (dataset per jam)
print("\nStatistik Deskriptif Dataset Per Jam (df_hour):")
display(df_hour.describe())
```

```python
# Melihat Distribusi Penyewaan Sepeda Berdasarkan Tahun & Bulan
df_rent_per_month = df_main.groupby(["yr", "mnth"]).agg(
    total_rent=("cnt", "sum"), 
    avg_rent=("cnt", "mean"), 
    median_rent=("cnt", "median")
).reset_index()

# Format angka agar lebih rapi
df_rent_per_month = df_rent_per_month.round(2)

# Menampilkan hasil
print("\nDistribusi Penyewaan Sepeda Berdasarkan Tahun & Bulan:")
display(df_rent_per_month)
```

```python
# Melihat Pengaruh Musim terhadap Penyewaan Sepeda
df_seasonal_rent = df_main.groupby("season").agg(
    total_rent=("cnt", "sum"),
    avg_rent=("cnt", "mean"),
    median_rent=("cnt", "median")
).reset_index()

# Mengubah angka kategori musim menjadi label yang lebih jelas
season_mapping = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
df_seasonal_rent["season"] = df_seasonal_rent["season"].replace(season_mapping)

# Format angka agar lebih rapi
df_seasonal_rent = df_seasonal_rent.round(2)

# Menampilkan hasil
print("\nPengaruh Musim terhadap Penyewaan Sepeda:")
display(df_seasonal_rent)
```

```python
# Melihat Pengaruh Cuaca terhadap Penyewaan Sepeda
df_weather_rent = df_main.groupby("weathersit").agg(
    total_rent=("cnt", "sum"),
    avg_rent=("cnt", "mean"),
    median_rent=("cnt", "median")
).reset_index()

# Mengubah angka kategori cuaca menjadi label yang lebih jelas
weather_mapping = {
    1: "Clear",
    2: "Mist",
    3: "Light Snow/Rain",
    4: "Heavy Rain/Snow"
}
df_weather_rent["weathersit"] = df_weather_rent["weathersit"].replace(weather_mapping)

# Format angka agar lebih rapi
df_weather_rent = df_weather_rent.round(2)

# Menampilkan hasil
print("\nPengaruh Cuaca terhadap Penyewaan Sepeda:")
display(df_weather_rent)
```

```python
# Melihat Tren Penyewaan Sepeda Berdasarkan Hari dalam Seminggu
df_weekday_rent = df_main.groupby("weekday").agg(
    total_rent=("cnt", "sum"),
    avg_rent=("cnt", "mean"),
    median_rent=("cnt", "median")
).reset_index()

# Mengubah angka kategori hari menjadi label nama hari
weekday_mapping = {
    0: "Senin", 1: "Selasa", 2: "Rabu", 3: "Kamis", 4: "Jumat",
    5: "Sabtu", 6: "Minggu"
}
df_weekday_rent["weekday"] = df_weekday_rent["weekday"].replace(weekday_mapping)

# Urutkan berdasarkan urutan hari dalam seminggu
order = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"]
df_weekday_rent["weekday"] = pd.Categorical(df_weekday_rent["weekday"], categories=order, ordered=True)
df_weekday_rent = df_weekday_rent.sort_values("weekday")

# Format angka agar lebih rapi
df_weekday_rent = df_weekday_rent.round(2)

# Menampilkan hasil
print("\nTren Penyewaan Sepeda Berdasarkan Hari dalam Seminggu:")
display(df_weekday_rent)
```

```python
# Analisis Hubungan Antara Hari Libur & Penyewaan Sepeda
df_holiday_rent = df_main.groupby("holiday").agg(
    total_rent=("cnt", "sum"),
    avg_rent=("cnt", "mean"),
    median_rent=("cnt", "median")
).reset_index()

# Mengubah angka kategori menjadi label
holiday_mapping = {0: "Bukan Libur", 1: "Hari Libur"}
df_holiday_rent["holiday"] = df_holiday_rent["holiday"].replace(holiday_mapping)

# Format angka agar lebih rapi
df_holiday_rent = df_holiday_rent.round(2)

# Menampilkan hasil
print("\nHubungan Antara Hari Libur & Penyewaan Sepeda:")
display(df_holiday_rent)
```

```python
# Menghitung korelasi antara suhu, kelembaban, kecepatan angin, dan penyewaan sepeda
corr_matrix = df_main[["temp", "hum", "windspeed", "cnt"]].corr().round(2)

# Menampilkan hasil korelasi dalam bentuk tabel
print("\nMatriks Korelasi Suhu, Kelembaban, Kecepatan Angin terhadap Penyewaan Sepeda:")
display(corr_matrix)
```

```python
# Statistik deskriptif casual & registered
print("\nStatistik Deskriptif untuk Pengguna Casual & Registered:")
display(df_main[["casual", "registered", "cnt"]].describe().round(2))

# Rata-rata penyewaan berdasarkan jenis pengguna di setiap hari dalam seminggu
df_main.groupby("weekday").agg({
    "casual": "mean",
    "registered": "mean"
})
```

### Pertanyaan 1: Bagaimana tren penyewaan sepeda sepanjang tahun?

```python
# Tren Penyewaan Sepeda Sepanjang Tahun

plt.figure(figsize=(12, 5))
sns.lineplot(data=df_main, x='mnth', y='cnt', hue='yr', marker='o', palette='coolwarm')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title("Tren Penyewaan Sepeda Sepanjang Tahun")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penyewaan")
plt.legend(title="Tahun", labels=["2011", "2012"])
plt.grid()
plt.show()
```

### Pertanyaan 2: Apakah ada pola penyewaan sepeda yang berbeda berdasarkan hari dan jam?

```python
# Pola Penyewaan Berdasarkan Hari dan Jamn 

plt.figure(figsize=(12, 6))

# Gunakan palet yang lebih jelas dan tambahkan marker
hourly_trend = df_main.groupby(["hr", "weekday"]).agg({"cnt": "mean"}).reset_index()
sns.lineplot(
    data=hourly_trend,
    x="hr",
    y="cnt",
    hue="weekday",
    palette="Set1",
    linewidth=2.5,
    marker="o"
)

# Tambahkan area highlight untuk jam sibuk
plt.axvspan(7, 9, color='gray', alpha=0.2, label="Jam Sibuk Pagi")
plt.axvspan(16, 18, color='gray', alpha=0.2, label="Jam Sibuk Sore")

plt.title("Tren Penyewaan Sepeda Berdasarkan Jam dalam Sehari untuk Setiap Hari")
plt.xlabel("Jam")
plt.ylabel("Jumlah Penyewaan")
plt.xticks(ticks=range(0, 24, 2))
plt.legend(title="Hari", labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.grid()
plt.show()
```

### Pertanyaan 4: Apakah ada perbedaan jumlah penyewaan antara hari kerja dan hari libur?

Musim dan cuaca berperan besar dalam tren penyewaan sepeda.
- Musim panas dan musim gugur memiliki tingkat penyewaan tertinggi, sedangkan musim dingin mengalami penurunan signifikan.
- Cuaca buruk seperti hujan dan salju berdampak negatif pada penyewaan, dengan jumlah peminjam yang turun drastis pada hari-hari dengan kondisi ekstrem.
- Suhu yang lebih hangat mendorong lebih banyak penyewaan, tetapi kelembaban tinggi dan kecepatan angin tinggi bisa menjadi faktor penghambat.

4. **Apakah ada perbedaan jumlah penyewaan antara hari kerja dan hari libur?**
- Hari kerja memiliki jumlah penyewaan lebih tinggi dibandingkan hari libur, terutama karena banyak pengguna yang memakai sepeda sebagai transportasi ke tempat kerja.
- Pada hari libur, penyewaan sepeda menurun, tetapi ada peningkatan signifikan dari pengguna casual, yang lebih cenderung menggunakan sepeda untuk aktivitas rekreasi.
- Pengguna registered lebih dominan pada hari kerja, yang menunjukkan bahwa mereka adalah komuter reguler.

5. **Bagaimana distribusi pengguna berdasarkan tipe (casual vs registered)?**
- Pengguna registered lebih mendominasi dibandingkan pengguna casual, terutama pada hari kerja.
- Pengguna casual lebih aktif di akhir pekan dan hari libur, sedangkan pengguna registered lebih sering menggunakan sepeda di hari kerja.
- Pengguna registered lebih konsisten sepanjang tahun, sedangkan pengguna casual cenderung lebih terpengaruh oleh musim dan cuaca.

## Conclusion

**Kesimpulan**

**1. Bagaimana tren penyewaan sepeda sepanjang tahun?**
- Penyewaan sepeda mengalami pola musiman yang jelas, dengan peningkatan signifikan pada bulan-bulan tertentu seperti Mei hingga Oktober dan penurunan pada bulan Januari dan Februari.
- Tahun 2011 dan 2012 menunjukkan pola yang serupa, dengan penyewaan tertinggi terjadi pada September dan Juni.
- Pola ini menunjukkan adanya faktor eksternal seperti cuaca dan musim yang mempengaruhi kebiasaan pengguna dalam menyewa sepeda.


**2. Apakah ada pola penyewaan sepeda yang berbeda berdasarkan hari dan jam?**
- Penyewaan sepeda lebih tinggi pada hari kerja dibandingkan hari libur, dengan puncak terjadi pada Rabu dan Kamis.
- Pada akhir pekan, hari Sabtu menunjukkan peningkatan penyewaan dibandingkan hari Minggu.
- Jam sibuk penyewaan terjadi pada pagi hari sekitar pukul 08.00 dan sore hari sekitar pukul 17.00, yang bertepatan dengan jam berangkat dan pulang kerja.
- Tren ini mengindikasikan bahwa sebagian besar pengguna menggunakan sepeda untuk keperluan transportasi harian.


**3. Bagaimana faktor eksternal seperti musim dan cuaca mempengaruhi jumlah penyewaan sepeda?**
- Penyewaan tertinggi terjadi pada musim panas (summer) dan gugur (fall), kemungkinan karena kondisi cuaca yang lebih kondusif untuk bersepeda.
- Penyewaan terendah terjadi pada musim semi (spring) dan terutama musim dingin (winter), terutama saat kondisi cuaca lebih ekstrem.
- Cuaca juga memainkan peran besar dalam pola penyewaan, di mana penyewaan menurun drastis saat hujan dan salju.
- Dengan kondisi cuaca cerah atau berawan, penyewaan tetap tinggi, menandakan preferensi pengguna terhadap kondisi cuaca yang nyaman.

**4. Apakah ada perbedaan jumlah penyewaan antara hari kerja dan hari libur?**
- Hari kerja memiliki jumlah penyewaan yang lebih tinggi dibandingkan dengan akhir pekan.
- Pada hari kerja, penyewaan didominasi oleh pengguna yang berangkat dan pulang kerja, terlihat dari puncak penyewaan di pagi dan sore hari.
- Sebaliknya, pada akhir pekan, pola penyewaan lebih merata sepanjang hari, mengindikasikan bahwa penggunaan sepeda lebih banyak untuk rekreasi dibandingkan transportasi rutin.

**5. Bagaimana jumlah penyewaan sepeda bervariasi di setiap musim dan kategori penggunaan?**
- Pengguna yang terdaftar (subscriber) lebih dominan dalam penyewaan sepeda pada hari kerja, sedangkan pengguna kasual (casual users) lebih banyak menyewa sepeda pada akhir pekan.
- Musim panas dan gugur menjadi waktu favorit bagi kedua kategori pengguna, tetapi pengguna kasual menunjukkan peningkatan lebih signifikan selama periode ini dibandingkan pengguna yang berlangganan.
- Pada musim dingin, jumlah penyewaan menurun drastis, baik untuk pengguna terdaftar maupun kasual, menandakan pengaruh besar dari faktor cuaca.


**Rekomendasi**

Penyewaan sepeda menunjukkan pola yang konsisten dipengaruhi oleh musim, cuaca, hari kerja, dan waktu dalam sehari. Untuk mengoptimalkan layanan dan meningkatkan penyewaan sepeda, operator dapat mempertimbangkan strategi berikut:
- Optimasi Ketersediaan Sepeda: Meningkatkan jumlah sepeda pada jam sibuk (pagi dan sore) serta selama musim panas dan gugur.
- Strategi Harga dan Promosi: Memberikan diskon atau promosi khusus pada musim dingin dan hari libur untuk meningkatkan jumlah penyewaan saat permintaan rendah.
- Perencanaan Infrastruktur: Menyediakan fasilitas tambahan seperti tempat parkir yang lebih banyak dan jalur sepeda yang aman untuk meningkatkan kenyamanan pengguna.
- Integrasi dengan Transportasi Publik: Menghubungkan penyewaan sepeda dengan sistem transportasi lain untuk menarik lebih banyak pelanggan, terutama pada hari kerja.

Dengan menerapkan strategi ini, operator dapat meningkatkan kepuasan pengguna dan memaksimalkan jumlah penyewaan sepanjang tahun.
